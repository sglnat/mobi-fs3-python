**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical $3 \times 3$ Prewitt filters:

```python
def prewitt_h(img):
    result_prewh = numpy.zeros(img.shape)
    n = (3 - 1) // 2
    prewh_matrix = numpy.array([[1,1,1], [0,0,0], [-1,-1,-1]]) #matrix andersrum!
    for p in numpy.ndindex(img.shape):
        if n<=p[0] and p[0]<img.shape[0]-n and n<=p[1] and p[1]<img.shape[1]-n:
            neighborhood_prewh = 0
            for q in numpy.ndindex(3, 3):
                neighborhood_intensity = img[p[0]+q[0]-n, p[1]+q[1]-n]
                neighborhood_prewh += (prewh_matrix[q[0], q[1]] * neighborhood_intensity)
            result_prewh[p[0], p[1]] = (neighborhood_prewh // 6)    
    return result_prewh

def prewitt_v(img):
    result_prewv = numpy.zeros(img.shape)
    n = (3 - 1) // 2
    prewv_matrix = numpy.array([[1,0,-1], [1,0,-1], [1,0,-1]])
    for p in numpy.ndindex(img.shape):
        if n<=p[0] and p[0]<img.shape[0]-n and n<=p[1] and p[1]<img.shape[1]-n:
            neighborhood_prewv = 0
            for q in numpy.ndindex(3, 3):
                neighborhood_intensity = img[p[0]+q[0]-n, p[1]+q[1]-n]
                neighborhood_prewv += (prewv_matrix[q[0], q[1]] * neighborhood_intensity)
            result_prewv[p[0], p[1]] = (neighborhood_prewv // 6)    
    return result_prewv
```
